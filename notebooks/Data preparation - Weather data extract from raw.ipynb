```python
import requests
import json
import datetime
from src import weather_repository
from src import repository_adapter
import json


    
result = repository_adapter.run_sql(
    """SELECT 
        raw_data,
        time
        FROM air.weather;
        ;""",
    [
        "raw_data", 
        "time", 
    ]
)


for index, weather in result.iterrows():
    raw_data = json.loads(weather['raw_data'])
    date = raw_data['data']['weather'][0]['date']
    for row in raw_data['data']['weather'][0]['hourly']:
        

            
        if(row['time'] == '0'):
            weather_repository.add_weather_measure_data(date + ' 00:00', row)
            weather_repository.add_weather_measure_data(date + ' 01:00', row)
            weather_repository.add_weather_measure_data(date + ' 02:00', row)
            
        if(row['time'] == '300'):
            weather_repository.add_weather_measure_data(date + ' 03:00', row)
            weather_repository.add_weather_measure_data(date + ' 04:00', row)
            weather_repository.add_weather_measure_data(date + ' 05:00', row)
            
        if(row['time'] == '600'):
            weather_repository.add_weather_measure_data(date + ' 06:00', row)
            weather_repository.add_weather_measure_data(date + ' 07:00', row)
            weather_repository.add_weather_measure_data(date + ' 08:00', row)
            
        if(row['time'] == '900'):
            weather_repository.add_weather_measure_data(date + ' 09:00', row)
            weather_repository.add_weather_measure_data(date + ' 10:00', row)
            weather_repository.add_weather_measure_data(date + ' 11:00', row)
            
        if(row['time'] == '1200'):
            weather_repository.add_weather_measure_data(date + ' 12:00', row)
            weather_repository.add_weather_measure_data(date + ' 13:00', row)
            weather_repository.add_weather_measure_data(date + ' 14:00', row)
            
        if(row['time'] == '1500'):
            weather_repository.add_weather_measure_data(date + ' 15:00', row)
            weather_repository.add_weather_measure_data(date + ' 16:00', row)
            weather_repository.add_weather_measure_data(date + ' 17:00', row)
            
        if(row['time'] == '1800'):
            weather_repository.add_weather_measure_data(date + ' 18:00', row)
            weather_repository.add_weather_measure_data(date + ' 19:00', row)
            weather_repository.add_weather_measure_data(date + ' 20:00', row)
            
        if(row['time'] == '2100'):
            weather_repository.add_weather_measure_data(date + ' 21:00', row)
            weather_repository.add_weather_measure_data(date + ' 22:00', row)
            weather_repository.add_weather_measure_data(date + ' 23:00', row)
```
